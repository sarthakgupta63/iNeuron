**Question 1.** Create a python program to find it out the Named entity recognition(NER) from the following sentence using spacy and visualizing those parts in sentence.

"I saw a kitten eating chicken in the kitchen"

```python
text = "I saw a kitten eating chicken in the kitchen"

doc = nlp(text)
for ent in doc.ents:
    print(f'Name: {ent.text},    Label: {ent.label_}')
```

```python
from spacy import displacy

displacy.serve(doc, style='ent')
```

**Question 2.** Find out the similar words from following list using spacy library.

["ship car truck motor-bike jeep hagskdshd"]

```python
txt = "ship car truck motor-bike jeep hagskdshd"
tokens = nlp(txt)
for i in range(len(tokens)):
    tok1 = tokens[i]
    for j in range(i+1, len(tokens)):
        tok2 = tokens[j]
        similarity = tok1.similarity(tok2)
        if similarity > 0.5:
            print(tok1, tok2, similarity)
```

**Question 3.** Visualize sentences including parts of speech(pos) tagging and linguistic annotation in it by using spacy library.

```python
import pandas as pd
from collections import defaultdict

pos_report = pd.DataFrame(columns=['Text', 'Lemma', 'Part_of_Speech', 'Tag', 
                                   'Dependency', 'Shape', 'Alphabet', 'Stop_Word'])

info = defaultdict(list)
for col in pos_report.columns:
    info[col] = list()

sent = '"The greatest glory in living lies not in never falling, but in rising every time we fall." - Nelson Mandela'

spacy_doc = nlp(sent)

for token in spacy_doc:
    info['Text'].append(token.text)
    info['Lemma'].append(token.lemma_)
    info['Part_of_Speech'].append(token.pos_)
    info['Tag'].append(token.tag_)
    info['Dependency'].append(token.dep_)
    info['Shape'].append(token.shape_)
    info['Alphabet'].append(token.is_alpha)
    info['Stop_Word'].append(token.is_stop)
    
for col in pos_report.columns:
    pos_report[col] = info[col]
    
pos_report
```
